```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
X = np.linspace(0, 10, 51)
```

```python
def Y(k, X):
    X = k * X
    return np.cos(X)
```

```python
plt.plot(X, Y(2, X))
plt.plot(X, Y(1, X))
plt.legend(labels=['K = 2', 'K = 1'])
```
